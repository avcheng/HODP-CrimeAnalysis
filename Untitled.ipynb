```python
import pandas as pd
import matplotlib.pyplot as plt


data1 = pd.read_csv("crime_data_log.csv")
```

```python
temp = {}

for index, row in data1.iterrows():
    if (row['Date'])[-1] == '9' and ((row['Date'])[0:2] == '01' or (row['Date'])[0:2] == '02' or (row['Date'])[0:2] == '03' or (row['Date'])[0:2] == '04'):
        if((row['Date'])[0:2] in temp):
            temp[(row['Date'])[0:2]] +=1
        else:
            temp[(row['Date'])[0:2]] =1
    else:
        if (row['Date'])[-1] == '8' and ((row['Date'])[0:2] == '05' or (row['Date'])[0:2] == '06' or (row['Date'])[0:2] == '07' or (row['Date'])[0:2] == '08' or (row['Date'])[0:2] == '09' or (row['Date'])[0:2] == '10' or (row['Date'])[0:2] == '11' or (row['Date'])[0:2] == '12'):
            if((row['Date'])[0:2] in temp):
                temp[(row['Date'])[0:2]] +=1
            else:
                temp[(row['Date'])[0:2]] =1

sd = sorted(temp.items())


x_val = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
y_val = [x[1] for x in sd]
```

```python
print x_val
print y_val
```

```python
plt.figure(figsize=(20,10))

plt.xlabel('Hour', fontsize = 20)
plt.ylabel('Number of Crimes', fontsize = 20)
plt.title('Number of Crimes Per Month', fontsize = 25)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.bar(x_val, y_val, color="#D2232A")
plt.show()
```
